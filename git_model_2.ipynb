## In this code im going to make a model for predict how much insurence can give after refer the client factors.

```python
# import part
import tensorflow as tf
from tensorflow.keras import layers
import pandas as pd
import numpy as np
from tensorflow.keras.callbacks import ModelCheckpoint
```

```python
# Read the csv file using pandas data frame
data_base = pd.read_csv("https://raw.githubusercontent.com/prvpitawala/ML_project_2_predict_insurance/main/data_base.csv")
data_base
```

```python
# lets one hot encode our data frame
encoded_data_base = pd.get_dummies(data_base)
encoded_data_base
```

```python
# create x and y data set
x = encoded_data_base.drop("charges", axis=1) # drop the chages part
y = encoded_data_base["charges"]
```

```python
print(x, y)
```

```python
# lets split our data to train and test part
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state=42)
```

```python
x_train, y_train
```

```python
# check the shape
x_train.shape, y_train.shape,  x_test.shape, y_test.shape
```

```python
# lets make the model

# set random seed
tf.random.set_seed(42)

# create the model
model = tf.keras.Sequential([
    layers.Dense(11, activation ="relu", input_shape =(None, 11)),
    layers.Dense(8, activation ="relu"),
    layers.Dense(4, activation ="relu"),
    layers.Dense(1)
])

# Compile the model
model.compile(loss = ["mae"],
              optimizer = tf.keras.optimizers.Adam(learning_rate = 0.001),
              metrics = ["mae"])

# Fit the model
model.fit(x_train,
          y_train,
          epochs = 2000,
          verbose = 1)
```

```python
# lets evaluate the model with the test data
model.evaluate(x_test, y_test)
```

```python
# lets test the predictions and real values
print(y_test[:5])
```

```python
model.predict(x_test[:5])
```
